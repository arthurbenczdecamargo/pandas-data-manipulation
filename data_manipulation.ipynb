```python
import pandas as pd
```

```python
#Read JSON file
```

```python
data = pd.read_json("info_properties.json")
data.head(10)
```

```python
# Normalize JSON file
```

```python
data = pd.json_normalize(data["info_properties"])
data.head(10)
```

```python
# Explode JSON file
```

```python
columns = list(data.columns)
columns
```

```python
data = data.explode(columns[3:])
data.head(10)
```

```python
# Reset index
```

```python
data.reset_index(inplace=True, drop=True)
data.head(10)
```

```python
data.info()
```

```python
import numpy as np
```

```python
# Dtype numeric conversion
```

```python
data["general_rate"] = data["general_rate"].astype(np.float64)
numeric_columns = ["max_guests", "total_restrooms", "total_rooms", "total_beds"]
data[numeric_columns] = data[numeric_columns].astype(np.int64)
data.info()
```

```python
# Clean price
```

```python
data["price"] = data["price"].apply(
    lambda x: x.replace("$", "").replace(",", "").strip()
)
data["price"] = data["price"].astype(np.float64)
data.info()
```

```python
# Clean fees
```

```python
data[["deposit_fee", "cleaning_fee"]] = data[["deposit_fee", "cleaning_fee"]].applymap(
    lambda x: x.replace("$", "").replace(",", "").strip()
)
data[["deposit_fee", "cleaning_fee"]] = data[["deposit_fee", "cleaning_fee"]].astype(
    np.float64
)
data.info()
```

```python
# Clean description
```

```python
data["local_description"] = (
    data["local_description"].str.lower().replace("[^a-zA-Z0-9\\-']", " ", regex=True)
)
data["local_description"] = data["local_description"].str.replace(
    "(?<!\\w)-(?!\\w)", " ", regex=True
)
data["local_description"].head(10)
```

```python
# Create tokens
```

```python
data["local_description"] = data["local_description"].str.split()
data["local_description"].head(10)
```

```python
data.head()
```

```python
# Clean included
```

```python
data["included"] = data["included"].str.replace('\\{|}|"', "", regex=True)
```

```python
# Create tokens
```

```python
data["included"] = data["included"].str.split(",")
data.head()
```
